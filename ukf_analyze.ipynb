```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np

my_cols=['sensor','nis']
with open('build/nis_data.txt') as f:
    table_data = pd.read_table(f, sep=' ', header=None, names=my_cols, lineterminator='\n')
```

```python
# find the radar and lidar data
r_indx = table_data['sensor'] == 'R'
r_nis = table_data['nis'][r_indx]
r_chi_square_95 = 7.81

l_indx = table_data['sensor'] == 'L'
l_nis = table_data['nis'][l_indx]
l_chi_square_95 = 5.99
plt.figure()
pts_nis,    = plt.plot(r_nis, 'bx')
thres_line, = plt.plot([0, len(r_nis) ], [r_chi_square_95, r_chi_square_95], 'r-', lw=2)
plt.legend([pts_nis, thres_line], ['NIS Values', '95% line']) 

plt.xlabel('Radar Measurement')
plt.ylabel('Radar Normalized Innovation Square')
plt.axis([0, len(r_nis), 0, 20])
plt.grid()
plt.savefig('radar_nis.png')
plt.show()

plt.figure()
pts_nis,    = plt.plot(l_nis, 'bx')
thres_line, = plt.plot([0, len(l_nis) ], [l_chi_square_95, l_chi_square_95], 'r-', lw=2)
plt.legend([pts_nis, thres_line], ['NIS Values', '95% line']) 
plt.xlabel('Lidar Measurement')
plt.ylabel('Lidar Normalized Innovation Square')
plt.axis([0, len(l_nis), 0, 20])
plt.grid()
plt.savefig('lidar_nis.png')
plt.show()
```
